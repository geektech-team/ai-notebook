# 12 理解线性回归

本节目标：理解线性模型、权重、偏置、均方误差和合成数据集。

## 1. 线性回归公式

二维特征的线性回归可以写成：

`y_hat = x1 * w1 + x2 * w2 + b`

其中：

- `x1, x2` 是两个输入特征。
- `w1, w2` 是权重。
- `b` 是偏置。
- `y_hat` 是模型预测。
- `y` 是真实标签。
- 误差来自噪声、模型能力不足、数据质量问题等。

```python
import torch
import matplotlib.pyplot as plt
from pathlib import Path

torch.manual_seed(42)
torch.set_printoptions(precision=4)

num_examples = 1000
true_w = torch.tensor([2.0, -3.4])
true_b = 4.2

X = torch.randn(num_examples, 2)
noise = torch.randn(num_examples) * 0.01
y_clean = X @ true_w + true_b
y = y_clean + noise

print("X shape:", tuple(X.shape))
print("true_w:", true_w)
print("true_b:", true_b)
print("y shape:", tuple(y.shape))
print("前 5 个样本特征:")
print(X[:5])
print("前 5 个标签:", y[:5])
```

```python
y_hat_from_true_params = X @ true_w + true_b
mse_from_true_params = ((y_hat_from_true_params - y) ** 2).mean()

print("用真实参数预测时的 MSE:", mse_from_true_params.item())
print("噪声均值:", noise.mean().item())
print("噪声标准差:", noise.std().item())
print("误差主要来自合成数据时加入的随机噪声。")
```

```python
assets_dir = Path("assets")
assets_dir.mkdir(exist_ok=True)
output_file = assets_dir / "12-linear-regression-data.png"

plt.figure(figsize=(6, 5))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, s=14, cmap="viridis")
plt.xlabel("feature x1")
plt.ylabel("feature x2")
plt.title("Synthetic Linear Regression Data")
plt.colorbar(scatter, label="label y")
plt.tight_layout()
plt.savefig(output_file, dpi=160)
plt.show()
print("数据可视化已保存:", output_file)
```

## 小结

线性回归假设标签可以由特征的加权和表示。训练的目标是从数据中恢复接近真实的 `w` 和 `b`。
