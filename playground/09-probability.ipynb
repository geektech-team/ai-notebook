# 09 概率基础

本节学习随机变量、概率分布、期望、方差、采样，并用 PyTorch 观察样本统计量如何接近理论值。

## 1. 最小概念

- 随机变量：结果不固定的变量，例如一次抽样得到的数字。
- 概率分布：描述随机变量可能取哪些值，以及每个值出现的可能性。
- 期望：随机变量的长期平均值，可以理解为理论平均数。
- 方差：随机变量围绕期望波动的程度，越大表示越分散。
- 采样：按照某个概率分布生成随机样本。

机器学习里很多事情都和概率有关：初始化参数、打乱数据、随机梯度下降、模型输出的不确定性。

```python
import torch

torch.manual_seed(42)
torch.set_printoptions(precision=4)
```

## 2. 从正态分布采样

我们使用一个正态分布：`N(2, 3^2)`。

- 理论均值：`2`
- 理论标准差：`3`
- 理论方差：`3^2 = 9`

样本数量越多，样本均值和样本方差通常越接近理论值。

```python
mean = 2.0
std = 3.0
theoretical_variance = std ** 2

sample_sizes = [10, 100, 1000, 10000]

print("理论均值:", mean)
print("理论方差:", theoretical_variance)
print()

for n in sample_sizes:
    samples = torch.normal(mean=mean, std=std, size=(n,))
    sample_mean = samples.mean()
    sample_variance = samples.var(unbiased=False)

    print(f"样本量 n = {n}")
    print("  samples shape:", tuple(samples.shape))
    print("  样本均值:", round(sample_mean.item(), 4))
    print("  样本方差:", round(sample_variance.item(), 4))
    print("  均值误差:", round(abs(sample_mean.item() - mean), 4))
    print("  方差误差:", round(abs(sample_variance.item() - theoretical_variance), 4))
    print()
```

## 3. 离散随机变量：伯努利分布

伯努利分布只有两个结果：`0` 或 `1`。

如果 `P(X=1)=p`，那么：

- 理论期望：`p`
- 理论方差：`p * (1 - p)`

这可以模拟一次点击、一次成功、一次正样本等二分类事件。

```python
p = 0.3
bernoulli = torch.distributions.Bernoulli(probs=p)
samples = bernoulli.sample((10000,))

sample_mean = samples.mean()
sample_variance = samples.var(unbiased=False)

print("伯努利分布 P(X=1):", p)
print("samples shape:", tuple(samples.shape))
print("前 20 个样本:", samples[:20])
print("理论期望:", p)
print("样本均值:", round(sample_mean.item(), 4))
print("理论方差:", round(p * (1 - p), 4))
print("样本方差:", round(sample_variance.item(), 4))
```

## 小结

- 随机变量的单次结果可能不稳定，但大量采样后的平均行为有规律。
- 期望描述长期平均水平，方差描述波动大小。
- 样本量越大，样本均值和样本方差通常越接近理论值。
- 机器学习训练中很多随机过程，都可以用概率视角理解。
