# 11 画出训练闭环

本节目标：把监督学习里的核心词串起来，画出“数据 -> 预测 -> 损失 -> 梯度 -> 参数更新”的训练闭环。

## 1. 最小概念

- 监督学习：用带标签的数据训练模型，让模型学会从特征预测标签。
- 特征：输入给模型的信息，通常记作 `X`。
- 标签：希望模型预测的真实答案，通常记作 `y`。
- 模型：把特征变成预测值的函数，通常记作 `y_hat = model(X)`。
- 损失函数：衡量预测值和真实标签差多少。
- 优化器：根据梯度更新参数，让损失变小。
- epoch：完整看完一遍训练集。
- batch：每次训练只取一小批数据，而不是一次看完整训练集。

````python
from pathlib import Path

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

assets_dir = Path("assets")
assets_dir.mkdir(exist_ok=True)
output_file = assets_dir / "11-training-loop-map.png"

nodes = [
    ("Data X, y", 0.08, 0.62),
    ("Prediction\n y_hat = model(X)", 0.30, 0.62),
    ("Loss\n loss(y_hat, y)", 0.54, 0.62),
    ("Backward\n loss.backward()", 0.78, 0.62),
    ("Gradient\n parameter.grad", 0.54, 0.26),
    ("Update\n optimizer.step()", 0.30, 0.26),
]

fig, ax = plt.subplots(figsize=(10, 4.8))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

box_w, box_h = 0.16, 0.16
for text, x, y in nodes:
    patch = FancyBboxPatch(
        (x, y), box_w, box_h,
        boxstyle="round,pad=0.02,rounding_size=0.025",
        linewidth=1.5,
        edgecolor="#2f4858",
        facecolor="#e7f2ef",
    )
    ax.add_patch(patch)
    ax.text(x + box_w / 2, y + box_h / 2, text, ha="center", va="center", fontsize=10)

arrows = [
    ((0.24, 0.70), (0.30, 0.70)),
    ((0.46, 0.70), (0.54, 0.70)),
    ((0.70, 0.70), (0.78, 0.70)),
    ((0.86, 0.62), (0.62, 0.42)),
    ((0.54, 0.34), (0.46, 0.34)),
    ((0.30, 0.42), (0.16, 0.62)),
]
for start, end in arrows:
    ax.add_patch(FancyArrowPatch(start, end, arrowstyle="->", mutation_scale=16, linewidth=1.5, color="#33658a"))

ax.text(0.5, 0.93, "Supervised Learning Training Loop", ha="center", va="center", fontsize=16, weight="bold")
ax.text(0.5, 0.08, "Each batch runs this loop; one epoch means all batches have been processed once.", ha="center", fontsize=10)
plt.tight_layout()
plt.savefig(output_file, dpi=160)
plt.show()
print("流程图已保存:", output_file)
````

## 2. Mermaid 版本

```mermaid
flowchart LR
    A["数据: 特征 X + 标签 y"] --> B["模型预测: y_hat = model(X)"]
    B --> C["损失: loss(y_hat, y)"]
    C --> D["反向传播: backward()"]
    D --> E["梯度: parameter.grad"]
    E --> F["参数更新: optimizer.step()"]
    F --> B
```

````python
terms = {
    "epoch": "完整遍历一遍训练集",
    "batch": "一次送进模型的一小批样本",
    "loss": "预测和标签之间的差距",
    "gradient": "告诉参数如何改变会让 loss 变大",
    "update": "沿梯度反方向调整参数，让 loss 变小",
}
for name, meaning in terms.items():
    print(f"{name}: {meaning}")
````

## 小结

训练不是一次计算，而是一个循环：模型先预测，损失函数衡量错多少，反向传播计算梯度，优化器用梯度更新参数。这个循环反复发生，模型才会逐渐变好。
