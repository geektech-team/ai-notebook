# 02 理解张量是什么

本节目标：理解标量、向量、矩阵、高维张量，以及张量的 `shape`、`dtype`、设备、索引和切片。

## 1. 最小概念

- 标量：0 维张量，一个数字。
- 向量：1 维张量，一排数字。
- 矩阵：2 维张量，像表格一样有行和列。
- 高维张量：3 维或更多维，可以表示一批图片、一段序列等。
- `shape`：每个维度的大小。
- `dtype`：张量中元素的数据类型。
- `device`：张量存放在哪里，例如 CPU、CUDA、MPS。

```python
import torch

torch.set_printoptions(precision=2)


def show(name, x):
    print(f"{name}:")
    print(x)
    print("shape:", tuple(x.shape))
    print("dtype:", x.dtype)
    print("device:", x.device)
    print()
```

## 2. 10 个张量小例子

```python
# 例子 1：标量，0 维张量
scalar = torch.tensor(3.14)
show("01 标量 scalar", scalar)
```

```python
# 例子 2：向量，1 维张量
vector = torch.tensor([1, 2, 3, 4])
show("02 向量 vector", vector)
```

```python
# 例子 3：矩阵，2 维张量
matrix = torch.tensor([[1, 2, 3], [4, 5, 6]])
show("03 矩阵 matrix", matrix)
```

```python
# 例子 4：高维张量，3 维张量
image_batch = torch.zeros((2, 3, 4))
show("04 高维张量 image_batch", image_batch)
print("可以理解为 2 个样本，每个样本是 3 x 4 的数据。")
```

```python
# 例子 5：常见创建方式
a = torch.arange(0, 12)
b = torch.ones((2, 3))
c = torch.randn((2, 3))

show("05-a arange 创建", a)
show("05-b ones 创建", b)
show("05-c randn 创建", c)
```

```python
# 例子 6：reshape 改变形状，不改变元素总数
x = torch.arange(12)
x_3_4 = x.reshape(3, 4)
x_2_2_3 = x.reshape(2, 2, 3)

show("06 原始 x", x)
show("06 reshape 成 3 x 4", x_3_4)
show("06 reshape 成 2 x 2 x 3", x_2_2_3)
```

```python
# 例子 7：索引和切片
x = torch.arange(12).reshape(3, 4)

show("07 原始 x", x)
print("第 0 行:", x[0])
print("第 1 行第 2 列:", x[1, 2])
print("所有行的第 1 列:", x[:, 1])
print("前 2 行，后 2 列:")
print(x[:2, 2:])
```

```python
# 例子 8：拼接
x = torch.arange(6).reshape(2, 3)
y = torch.arange(6, 12).reshape(2, 3)

concat_rows = torch.cat([x, y], dim=0)
concat_cols = torch.cat([x, y], dim=1)

show("08 x", x)
show("08 y", y)
show("08 按行拼接 dim=0", concat_rows)
show("08 按列拼接 dim=1", concat_cols)
```

```python
# 例子 9：求和和均值
x = torch.arange(12, dtype=torch.float32).reshape(3, 4)

show("09 x", x)
print("所有元素求和:", x.sum())
print("所有元素均值:", x.mean())
print("每一列求和 dim=0:", x.sum(dim=0))
print("每一行求均值 dim=1:", x.mean(dim=1))
```

```python
# 例子 10：dtype 和 device
int_tensor = torch.tensor([1, 2, 3], dtype=torch.int64)
float_tensor = int_tensor.float()

device = "mps" if getattr(torch.backends, "mps", None) is not None and torch.backends.mps.is_available() else "cpu"
device_tensor = float_tensor.to(device)

show("10 整数张量", int_tensor)
show("10 转成 float", float_tensor)
show("10 移动到可用设备", device_tensor)
```

## 小结

张量是深度学习中最基本的数据结构。学习张量时，优先养成三个习惯：看 `shape`、看 `dtype`、看 `device`。
